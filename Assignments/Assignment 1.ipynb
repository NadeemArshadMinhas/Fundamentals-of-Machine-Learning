```python
import seaborn as sns # importing seaborn package for ploting
import pandas as pd # importing pandas for abbreviation in commands
import matplotlib.pyplot as plt # importing matplotlip for ploting graphs
```

```python
# importing 2 datasets as in csv format
steps = pd.read_csv("steps data.csv", delimiter =";")
survey = pd.read_csv("survey data.csv")
# visualizing both data sets for data analysis
steps.head(10)
```

```python
survey.head(10)
```

```python
#combining the 2 datasets as whole in dataframe
df = pd.merge(steps, survey, on ="id")
df.head(10)
```

```python
# Getting information of variables in dataframe
df.info()
```

```python
df["weight"].value_counts()
```

```python
# ploting a plot for the weight among the ids
sns.distplot(df["weight"].dropna(), kde=False)
plt.title("weight of ids")
plt.xlabel("weight(kg)")
plt.show()
```

```python
# Weight statistics in mean and median
# Labeling the plot
df['weight'].name = 'weight' 
height = df['weight'].dropna() #Selecting colum size .
sns.distplot(height, kde=False) 

mean_size = df['weight'].mean()
median_size = df['weight'].median()

print(f'The median weight is: {median_size} kg')
print(f'The mean weight is: {mean_size} kg')

sns.distplot(height, kde=False)
plt.title('weight of FitBit users')
plt.axvline(median_size, 0, 100, color='green', label='median') 
plt.axvline(mean_size, 0, 100, color='red', label='mean') 
plt.legend() # adding a legend.
plt.show()
```

```python
# boxplot for weight distribution
sns.boxplot(df["weight"])
plt.title('weight of ids')
plt.xlabel('weight in kg')
plt.show()
```

```python
# density plot of weight
sns.distplot(height, rug=True)
plt.title('weight distribution')
plt.xlabel('weight in kg')
plt.ylabel('Amount of users')
plt.show()
```

```python
# Overall statistics for weight variable 

print('median: ' + str(float(df[['weight']].median())))
print('mode: ' + str(df[['weight']].mode()['weight'][0]))
print('mean: ' + str(float(df[['weight']].mean())))
print('standard deviation: ' + str(float(df[['weight']].std())))
print('variance: ' + str(float(df[['weight']].var())))
```

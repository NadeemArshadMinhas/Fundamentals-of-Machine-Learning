## Assignment 4 - Speed Dating 


Precicting the variable dec_o with Kaggle Data Set by K-Nearest Neighbor Algorithm (KNN). The k-nearest neighbors (KNN) algorithm is a simple, supervised machine learning algorithm that can be used to solve both classification and regression problems. 

```python
# Importing all Required packages for the assignment
import seaborn as sns
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Chosing Speed dating Experiment data Set

df = pd.read_csv('Speed Dating Data.csv', encoding = 'unicode_escape')
df.head(10)
```

```python
# Generating the subset and droping NA's
# Selecting 8 variables with strong predictive value
df_subset = df[['dec_o', 'age', 'samerace', 'fun_o', 'match', 'attr_o', 'like_o', 'prob_o']]
df_subset = df_subset.dropna()

df_subset.head() #The head()of the resulting data frame
```

```python
df_subset.corr() # checking the co-relation int the resulting data set
```

The output shows co-relations among variables , so we can apply K-Nearest Neighbour Algorithm by first spliting the data set into training and test set .

```python

X = df_subset[['age', 'samerace', 'fun_o', 'match', 'attr_o', 'like_o', 'prob_o']] #create the X matrix
y = df_subset['dec_o'] #create the y-variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1) #split the data, store it into different variables
```

```python

from sklearn.neighbors import KNeighborsClassifier # Object class

knn = KNeighborsClassifier(n_neighbors = 6) # KNN-classifier with 5 neighbors (default)
knn = knn.fit(X_train, y_train) # Fits the k-nearest neigbor model 
knn.score(X_test, y_test)
```

The accuracy shows the Model around 77% .

```python

from sklearn.metrics import confusion_matrix
y_test_pred = knn.predict(X_test) #the predicted values
cm = confusion_matrix(y_test, y_test_pred) #creates a "confusion matrix"
cm
```

```python
conf_matrix = pd.DataFrame(cm, index=['Match (actual)', 'No match (actual)'], columns = ['Match (predicted)', 'No match (predicted)']) 
# conf matrix, columns are always the predicted values
conf_matrix
```
